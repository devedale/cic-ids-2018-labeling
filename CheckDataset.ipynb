```python
import pandas as pd

def analyze_labels(file_path, label_col="Label", chunksize=200000):
    total_rows = 0
    label_counts = {}

    for chunk in pd.read_csv(
        file_path,
        usecols=[label_col],
        chunksize=chunksize,
        low_memory=False
    ):
        total_rows += len(chunk)
        vc = chunk[label_col].value_counts(dropna=False)

        for label, count in vc.items():
            label_counts[label] = label_counts.get(label, 0) + count

    return total_rows, label_counts


pairs = {
    "Thursday-01-03-2018": (
        "./data/Processed Traffic Data for ML Algorithms/Thursday-01-03-2018_TrafficForML_CICFlowMeter.csv",
        "./data/CSECICIDS2018_improved/Thursday-01-03-2018.csv"
    ),
    "Friday-02-03-2018": (
        "./data/Processed Traffic Data for ML Algorithms/Friday-02-03-2018_TrafficForML_CICFlowMeter.csv",
        "./data/CSECICIDS2018_improved/Friday-02-03-2018.csv"
    ),
    "Wednesday-14-02-2018": (
        "./data/Processed Traffic Data for ML Algorithms/Wednesday-14-02-2018_TrafficForML_CICFlowMeter.csv",
        "./data/CSECICIDS2018_improved/Wednesday-14-02-2018.csv"
    )
}


for day, (file1, file2) in pairs.items():

    print("\n" + "="*70)
    print(f"DATASET: {day}")
    print("="*70)

    rows1, labels1 = analyze_labels(file1)
    rows2, labels2 = analyze_labels(file2)

    print("\n--- ORIGINAL (CIC-IDS-2018-ML-DATA) ---")
    print("Total rows:", rows1)
    for k, v in sorted(labels1.items()):
        print(f"  {k}: {v}")

    print("\n--- IMPROVED (Kaggle) ---")
    print("Total rows:", rows2)
    for k, v in sorted(labels2.items()):
        print(f"  {k}: {v}")

    print("\n--- CONFRONTO ---")
    print("Differenza numero record (original - improved):", rows1 - rows2)

    labels_only_1 = set(labels1.keys()) - set(labels2.keys())
    labels_only_2 = set(labels2.keys()) - set(labels1.keys())

    print("Label dataset original:", labels_only_1)
    print("Label dataset improved(kaggle):", labels_only_2)
```

I dati di CIC-IDS-2018 "Processed Traffic Data for ML Algorithms" scaricati da aws ufficiali gia etichettati non hanno dest IP, src IP e src Port quindi ip2vec si puo fare ma sarebbe sicuramente ridimensionato di molto analizzando soltanto **frasi di due token** **protocollo** e **dst port**.

La porta di destinazione e il protocollo effettivamente sono informazioni cruciali nel rilevare il tipo di iterazione la **porta d'origine e gli IP andrebbero semplicemente ad overfittare il dataset** in un scenario simulato

La **porta di origine** è spesso una porta **effimera**, selezionata dallo stack TCP/IP da un intervallo di porte dinamiche.
Questa viene assegnata con lo scopo di **gestire in maniera univoca piu connessioni in parallelo**.
Poiché queste porte vengono assegnate dinamicamente e spesso con meccanismi pseudo-casuali, tendono a contenere poca informazione semantica sul tipo di servizio o di applicazione, risultando quindi spesso poco indicative per l’analisi del traffico.

Addestrare sugli IP di una qualsiasi simulazione di fatto sarebbe poco sensato sempre. **Diverso sarebbe uno scenario reale** nel quale si potrebbero di fatto cogliere delle **relazioni fra IP della rete globale tramite ip2vec**.
