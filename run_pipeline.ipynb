## 3. Run the Pipeline

On Google Colab, `source .venv/bin/activate` is not available (the venv step is skipped
by `setup.sh`), so the system Python interpreter is used instead. On a local machine
`.venv/bin/python` is used. In both cases the behaviour is equivalent to
`source .venv/bin/activate && python main.py`.


````python
import subprocess, sys, os
repo = "cic-ids-2018-labeling"
python = os.path.join(repo, ".venv", "bin", "python") if os.path.exists(os.path.join(repo, ".venv")) else sys.executable
result = subprocess.run([python, "main.py"], cwd=repo)
if result.returncode != 0:
    raise RuntimeError(f"main.py exited with code {result.returncode}")
````
